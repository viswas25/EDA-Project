# Extracting Product_Name Column

```python
product_names = []

for page in range(1, 20):  # scrape 19 pages (~450+ products)
    url = f"https://www.flipkart.com/search?q=chimneys&page={page}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    products = soup.find_all("div", {"class": "KzDlHZ"})
    print(f"Page {page}: Found {len(products)} product names")

    for p in products:
        name = p.get_text(strip=True)
        product_names.append(name)
    
    time.sleep(2)  # polite scraping delay

print("✅ Total product names:", len(product_names))
print(product_names[:10])
```

# Rating Column

```python
ratings = []

for page in range(1, 22):
    url = f"https://www.flipkart.com/search?q=chimneys&page={page}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    cards = soup.select("div._75nlfW, div._1sdMkc")
    print(f"Page {page}: {len(cards)} product blocks")

    for c in cards:
        rating_tag = c.select_one("div.XQDdHH")
        # append either the rating or None every time
        rating = rating_tag.get_text(strip=True) if rating_tag else None
        ratings.append(rating)

    time.sleep(1)

print("✅ Total Ratings collected:", len(ratings))
```

# Review Count Column

```python
pages = range(1, 20)

review_counts = []

for page in pages:
    url = f"https://www.flipkart.com/search?q=chimneys&page={page}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    cards = soup.select("div._75nlfW, div._1sdMkc")
    print(f"Page {page}: {len(cards)} product blocks")

    for c in cards:
        review_count = None
        # find any <span> containing "Reviews"
        spans = c.find_all("span")
        for s in spans:
            if "Reviews" in s.get_text():
                # extract only number
                match = re.search(r"(\d[\d,]*)\s*Reviews", s.get_text())
                if match:
                    review_count = match.group(1).replace(",", "")
                    break
        review_counts.append(review_count)

    time.sleep(1)

print("✅ Total Review Counts:", len(review_counts))
print(review_counts[:15])
```

# Suction Capacity Column

```python
pages = range(1, 20)

suction_capacity = []

for page in pages:
    url = f"https://www.flipkart.com/search?q=chimneys&page={page}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    cards = soup.select("div._75nlfW, div._1sdMkc")
    print(f"Page {page}: {len(cards)} product blocks")

    for c in cards:
        suction = None
        # Escape the '+' in class names => .J\+igdf
        for li in c.select("li.J\\+igdf, li.rgWa7D, li._1nqp6V"):
            text = li.get_text(strip=True)
            if "Suction" in text or "m³/hr" in text or "CMH" in text:
                match = re.search(r"(\d[\d,]*)\s*(m³/hr|CMH)", text)
                suction = match.group(0).replace(",", "") if match else text
                break
        suction_capacity.append(suction)

    time.sleep(1)

print("✅ Total Suction Capacities:", len(suction_capacity))
print(suction_capacity[:20])
```

# Size Column

```python
pages = range(1, 20)  # ~456 rows total

sizes = []

for page in pages:
    url = f"https://www.flipkart.com/search?q=chimneys&page={page}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    # Each full product card
    cards = soup.select("div._75nlfW, div._1sdMkc")
    print(f"Page {page}: {len(cards)} product blocks")

    for c in cards:
        size = None
        # Escape '+' in class selector -> .J\+igdf
        for li in c.select("li.J\\+igdf, li.rgWa7D, li._1nqp6V"):
            text = li.get_text(strip=True)
            if "Size" in text and "cm" in text:
                # Extract only the numeric size with "cm"
                match = re.search(r"(\d+)\s*cm", text, flags=re.I)
                size = match.group(1) + " cm" if match else text
                break
        sizes.append(size)

    time.sleep(1)

print("✅ Total Sizes:", len(sizes))
print(sizes[:20])
```

# Offer Percentage

```python
import requests, time, re
from bs4 import BeautifulSoup

headers = {"User-Agent": "Mozilla/5.0"}
pages = range(1, 22)

offers = []

for page in pages:
    url = f"https://www.flipkart.com/search?q=chimneys&page={page}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    # select each product block
    cards = soup.select("div._75nlfW, div._1sdMkc")
    print(f"Page {page}: {len(cards)} product blocks")

    for c in cards:
        offer = None
        
        # find <span> containing '% off' or 'Offer'
        for s in c.find_all("span"):
            text = s.get_text(strip=True)
            if re.search(r"%\s*off|Offer", text, flags=re.I):
                offer = text
                break

        offers.append(offer)

    time.sleep(1)

print("✅ Total Offers:", len(offers))
print(offers[:20])
```

# Instead of reducing each row size after extraction i have reduced in once to 456

```python
target_len = 456  # final dataset size

def fix_length(lst, target_len=456):
    """Ensure each list has exactly target_len elements."""
    if len(lst) < target_len:
        lst += [None] * (target_len - len(lst))  # pad with None
    elif len(lst) > target_len:
        lst = lst[:target_len]  # trim extra
    return lst

# Apply to all your column lists
product_names = fix_length(product_names, target_len)
brands = fix_length(brands, target_len)
prices = fix_length(prices, target_len)
ratings = fix_length(ratings, target_len)
offers = fix_length(offers, target_len)
review_counts = fix_length(review_counts, target_len)
suction_capacity = fix_length(suction_capacity, target_len)
sizes = fix_length(sizes, target_len)


```

# Now safely combine all lists into one aligned DataFrame

```python
import pandas as pd

df = pd.DataFrame({
    "Product Name": product_names,
    "Brand": brands,
    "Price": prices,
    "Rating": ratings,
    "Offer": offers,
    "Review Count": review_counts,
    "Suction Capacity": suction_capacity,
    "Size": sizes
})

print("✅ Final DataFrame Shape:", df.shape)
df.head()
```
